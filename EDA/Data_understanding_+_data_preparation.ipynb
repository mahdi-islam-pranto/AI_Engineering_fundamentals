```python
# drop by index and make a copy dataset (for too many column dropping)
df = df[['coaster_name', 'Length', 'Speed', 'Location', 'Status', 'Opening date',
       'Type', 'Manufacturer', 'Height restriction', 'Model', 'Height',
       'Inversions',
         #'Lift/launch system', 'Cost', 'Trains', 'Park section',
       'Duration',
         'Capacity',
         #'G-force', 'Designer', 'Max vertical angle',
       'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       'Closing date', 'Opened',
         #'Replaced by', 'Website',
       'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       #'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 'height_value', 'height_unit', 'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.shape
```

```python
df.info()
```

```python
# change the d-type of a collumn
df["opening_date_clean"] = pd.to_datetime(df["opening_date_clean"])
df['opening_date_clean']
```

```python
# Rename Columns
df = df.rename(columns={
    "coaster_name": "Coaster_Name",
    "Opening date": "Opening_Date",
    "Height restriction": "Height_Restriction",
    "Max vertical angle": "Max_Vertical_Angle",
    "Closing date": "Closing_Date",
    "Flash Pass Available": "Flash_Pass_Available",
    "Must transfer from wheelchair": "Must_Transfer_From_Wheelchair",
    "Fastrack available": "Fastrack_Available",
    "Soft opening date": "Soft_Opening_Date",
    "Flash Pass available": "Flash_Pass_Available",

})
```

```python
# check how many missing values are there
df.isna().sum()
```

```python
# check for duplicate rows within a column
df.duplicated(subset=['Coaster_Name', 'Opening_Date']).sum()
```

```python
# keep only unique rows / non duplicates
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Opening_Date'])].reset_index(drop=True).copy()
```
