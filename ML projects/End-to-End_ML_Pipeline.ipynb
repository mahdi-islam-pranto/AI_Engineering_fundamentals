## üîÅ What is a Machine Learning Pipeline?

A **machine learning pipeline** is a series of steps that help us go from raw data to a trained model ready for prediction.

### ‚ú® Typical Steps in a Pipeline
1. Data Collection ‚Äì Get the data (e.g., CSV files, databases, APIs)
2. Data Cleaning & Preprocessing ‚Äì Handle missing values, convert types, etc.
3. Feature Engineering ‚Äì Create useful input features for the model
4. Splitting the Data ‚Äì Train-test split
5. Model Selection & Training ‚Äì Choose and train a model
6. Model Evaluation ‚Äì Check performance (accuracy, precision, etc.)
7. Model Tuning ‚Äì Improve it using techniques like cross-validation
8. Deployment ‚Äì Use the model in a real-world app (advanced)

## üìâ Underfitting vs Overfitting

**Underfitting**
- Model is too simple
- Poor performance on both training and test data

**Overfitting**
- Model is too complex
- Great on training data, bad on test data

```python
import numpy as np
import matplotlib.pyplot as plt

# Dummy training data
np.random.seed(1)
rooms = np.array([1, 2, 3, 4, 5, 6, 7, 8])
prices = 50000 + 50000 * rooms + np.random.normal(0, 20000, size=rooms.shape)

# Train a simple linear regression model (just fit a line)
coeffs = np.polyfit(rooms, prices, deg=1)
regression_line = np.poly1d(coeffs)

# New data point to predict
new_room = 6.5
predicted_price = regression_line(new_room)

# Plotting
plt.figure(figsize=(10, 6))

# Actual data
plt.scatter(rooms, prices, color='black', label='Training Data', s=60)

# Regression line
x_vals = np.linspace(1, 9, 100)
plt.plot(x_vals, regression_line(x_vals), color='green', label='Learned Regression Line')

# New point prediction
plt.scatter([new_room], [predicted_price], color='red', label='New Point (Predicted)', s=100)
plt.vlines(new_room, ymin=0, ymax=predicted_price, linestyles='dotted', colors='red')
plt.hlines(predicted_price, xmin=0, xmax=new_room, linestyles='dotted', colors='red')

# Annotations
plt.text(new_room + 0.2, predicted_price, f"Predicted Price ‚âà {int(predicted_price):,} BDT", color='red')

# Labels and Legend
plt.title("Linear Regression: Predicting Price for New Data Point")
plt.xlabel("Number of Rooms")
plt.ylabel("Price (BDT)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dummy data: rooms vs price
np.random.seed(42)
rooms = np.linspace(1, 10, 30)
true_prices = 50_000 + rooms * 50_000 + np.random.normal(0, 20_000, size=rooms.shape)

# Simulated predictions
underfit = np.full_like(rooms, np.mean(true_prices))  # Constant prediction
good_fit = 50_000 + rooms * 50_000                    # Linear
overfit = true_prices  # Make the overfit equation perfectly fit the true data

# Plotting
plt.figure(figsize=(12, 6))

plt.scatter(rooms, true_prices, color='black', label='True Data', s=60)

plt.plot(rooms, underfit, label='Underfit Model', linestyle='--', color='red')
plt.plot(rooms, good_fit, label='Good Fit Model', linestyle='-', color='green')
plt.plot(rooms, overfit, label='Overfit Model', linestyle='-.', color='blue')

plt.title("Underfit vs Good Fit vs Overfit (Dummy House Price Data)")
plt.xlabel("Number of Rooms")
plt.ylabel("Price (BDT)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## üß† Bias and Variance

**Bias**
- Error from wrong assumptions (e.g., linear model for nonlinear data)
- Leads to **underfitting**

**Variance**
- Error from model being too sensitive to small changes
- Leads to **overfitting**

**Bias-Variance Table:**
| Concept      | Bias            | Variance        |
|--------------|------------------|------------------|
| Too Simple   | ‚úÖ High Bias     | ‚ùå Low Variance |
| Too Complex  | ‚ùå Low Bias      | ‚úÖ High Variance|
| Just Right   | ‚úÖ Low Bias      | ‚úÖ Low Variance |
